## **Задача: “Общие друзья в ВК”**

**Описание:** Определите общих друзей между двумя пользователями в ВК.

**Что нужно сделать:** 
- Запросите список друзей первого пользователя (через запятую) 
- Запросите список друзей второго пользователя (через запятую)
- Преобразуйте в множества для удаления дубликатов
- Найдите общих друзей (пересечение множеств)
- Покажите уникальных друзей каждого пользователя
- Покажите общих друзей и их количество
  
**Входные данные:** строки с именами через запятую<br>
**Выходные данные:** множества и операции над ними

**Пример работы программы:**

ПОИСК ОБЩИХ ДРУЗЕЙ В СОЦСЕТЯХ

Введите друзей пользователя 1 (через запятую): Анна,Борис,Виктор,Дарья,Елена<br>
Введите друзей пользователя 2 (через запятую): Борис,Григорий,Дарья,Жанна,Зоя

=== РЕЗУЛЬТАТЫ АНАЛИЗА ===

Друзья пользователя 1: 5 человек<br>
Друзья пользователя 2: 5 человек <br>
Общие друзья: Борис, Дарья<br>
Количество общих друзей: 2

Уникальные друзья user1: Анна, Виктор, Елена<br>
Уникальные друзья user2: Григорий, Жанна, Зоя

```python
print("ПОИСК ОБЩИХ ДРУЗЕЙ В СОЦСЕТЯХ\n")

friends_user1 = input("Введите друзей пользователя 1 (через запятую):")
friends_user2 = input("Введите друзей пользователя 2 (через запятую):")

set_user1 = set(friends_user1.split(','))
set_user2 = set(friends_user2.split(','))

print("\n===РЕЗУЛЬТАТЫ АНАЛИЗА===\n")


print (f"Друзья пользователя 1: {len(set_user1)}")
print (f"Друзья пользователя 2: {len(set_user2)}")

intersection_set = set_user1.intersection(set_user2)
length = len(intersection_set)
print(f"Общие друзья: {intersection_set}")
print(f'Количество общих друзей: {length}')


unique_user1 = set_user1 - intersection_set
unique_user2 = set_user2 - intersection_set

print(f"\nУникальные друзья пользователя 1: {unique_user1}")
print(f"Уникальные друзья пользователя 2: {unique_user2}")
```

# **Кортежи (tuple)**
# **Задача: “Расчёт расстояния”**

**Описание:** Рассчитайте расстояние между двумя точками.

**Что нужно сделать:** 
- Ввести координаты первой точки (широта, долгота)
- Ввести координаты второй точки (широта, долгота)
- Сформируйте кортежи координат и выведите их на экран
- Рассчитайте расстояние между двумя точками и выведите их на экран

**Входные данные:** числа (координаты точек)<br>
**Выходные данные**: кортежи и числа

**Пример работы программы:**

РАСЧЕТ РАССТОЯНИЯ МЕЖДУ ТОЧКАМИ

Введите координаты точки A:<br>
Широта: 55.7558<br>
Долгота: 37.6173<br>

Введите координаты точки B:<br>
Широта: 59.9343<br>
Долгота: 30.3351

=== РЕЗУЛЬТАТЫ ===<br>
Точка A: (55.7558, 37.6173)<br>
Точка B: (59.9343, 30.3351)

Расстояние между точками: 634.2

```python
import math

print("\nРАСЧЕТ РАССТОЯНИЯ МЕЖДУ ТОЧКАМИ\n\nВведите координаты точки А:")

lat_a = float(input("Широта:"))
lon_a = float(input("Долгота:"))

print("\nВведите координаты точки B:")
lat_b = float(input("Широта:"))
lon_b = float(input("Долгота:"))

point_a = (lat_a, lon_a)
point_b = (lat_b, lon_b)

print("\n===РЕЗУЛЬТАТЫ===\n")
print(f"Точка А: {point_a}")
print(f"Точка В: {point_b}")

R = 6371  
lat1, lon1 = map(math.radians, point_a)
lat2, lon2 = map(math.radians, point_b)

dlon = lon2 - lon1
dlat = lat2 - lat1
a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
c = 2 * math.asin(math.sqrt(a))

distance = R * c

print(f"\nРасстояние между точками: {distance}")
```

# **Словари (dict)**
## **Задача: “Кулинарная книга рецептов”**

**Описание:** Создайте кулинарную книгу с простыми рецептами и их ингредиентами.

**Что нужно сделать:**
- Создать свой словарь для каждого рецепта, в котором содержится информация о названии, ингредиентам, времени приготовления
- Вывести на экран список всех рецептов
- Показать детальную информацию о каждом рецепте
- Найти самый быстрый и самый долгий рецепт

**Пример работы программы:**

МОЯ КУЛИНАРНАЯ КНИГА

=== ВСЕ РЕЦЕПТЫ ===<br>
1. Блины<br>
2. Салат Цезарь<br>
3. Шоколадное печенье

=== ДЕТАЛИ РЕЦЕПТОВ ===<br>
Блины:
- Ингредиенты: мука, яйца, молоко, сахар, масло
- Время: 30 мин

Салат Цезарь:
- Ингредиенты: салат, курица, сухарики, сыр, соус
- Время: 20 мин

Шоколадное печенье:
- Ингредиенты: мука, шоколад, масло, сахар, яйца
- Время: 45 мин

Самый быстрый: Салат Цезарь (20 мин)
Самый долгий: Шоколадное печенье (45 мин)


```python
recipes = [
    {
        "name": "Блины",
        "ingredients": ["мука", "яйца", "молоко", "сахар", "масло"],
        "cooking_time": 30
    },
    {
        "name": "Салат Цезарь",
        "ingredients": ["салат", "курица", "сухарики", "сыр", "соус"],
        "cooking_time": 20
    },
    {
        "name": "Шоколадное печенье",
        "ingredients": ["мука", "шоколад", "масло", "сахар", "яйца"],
        "cooking_time": 45
    }
]

print("МОЯ КУЛИНАРНАЯ КНИГА\n")
print("=== ВСЕ РЕЦЕПТЫ ===")
for recipe in recipes:
    print(recipe["name"])

print("\n=== ДЕТАЛИ РЕЦЕПТОВ ===")
for recipe in recipes:
    print(f"{recipe['name']}:\n")
    print(f"Ингредиенты: {', '.join(recipe['ingredients'])}")
    print(f"Время: {recipe['cooking_time']} мин\n")


fastest_recipe = min(recipes, key=lambda x: x["cooking_time"])
slowest_recipe = max(recipes, key=lambda x: x["cooking_time"])

print(f"Самый быстрый: {fastest_recipe['name']} ({fastest_recipe['cooking_time']} мин)")
print(f"Самый долгий: {slowest_recipe['name']} ({slowest_recipe['cooking_time']} мин)")
```

**Что нужно сделать:** 
- Попросить пользователя ввести команду и текст
- Использовать оператор case для обработки разных команд
Доступные команды:
- uppercase: преобразовать в верхний регистр
- lowercase: преобразовать в нижний регистр
- length: посчитать длину текста
- words: посчитать количество слов
- reverse: перевернуть текст
- Выполнить соответствующую операцию с текстом
- Вывести результат
  
**Пример работы программы:**

УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ

Доступные команды:
- uppercase: преобразовать в верхний регистр
- lowercase: преобразовать в нижний регистр  
- length: посчитать длину текста
- words: посчитать количество слов
- reverse: перевернуть текст

Введите команду: length<br>

Дополнительный пример работы:<br>

Результат: olleh

Введите команду: uppercase<br>
Введите текст: привет мир

Результат: ПРИВЕТ МИР


```python
print("УМНЫЙ КАЛЬКУЛЯТОР ОПЕРАЦИЙ С ТЕКСТОМ\n")
print("Доступные команды:")
print("uppercase: преобразовать в верхний регистр")
print("lowercase: преобразовать в нижний регистр")
print("length: посчитать длину текста")
print("words: посчитать количество слов")
print("reverse: перевернуть текст\n")

command = input("Введите команду: ").strip()
text = input("Введите текст: ").strip()

if command == "uppercase":
    result = text.upper()
elif command == "lowercase":
    result = text.lower()
elif command == "length":
    result = f"Длина текста: {len(text)} символов"
elif command == "words":
    word_count = len(text.split())
    result = f"Количество слов: {word_count}"
elif command == "reverse":
    result = text[::-1]
else:
    result = "Неизвестная команда"

print(f"Результат: {result}")
```
